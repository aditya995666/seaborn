```python
df=pd.read_csv("C:\\Users\\DELL\\OneDrive\\Desktop\\Python_Test\\Who_is_responsible_for_global_warming.csv")
df.head()
```

```python
df_1=df.drop(columns=["Country Code","Indicator Name","Indicator Code"],axis=1).set_index("Country Name")
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df_1)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df_1,vmin=False)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df_1,vmin=False,cmap="coolwarm")
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df_1,vmin=False,cmap="coolwarm",annot=True)
```

```python
plt.figure(figsize=(16,9))
annot_kws={"fontsize":10,"fontstyle":"italic","color":"k",
           "alpha":0.9,
           "backgroundcolor":"w"
           
           
          }
sns.heatmap(df_1,vmin=False,cmap="coolwarm",annot=True,annot_kws=annot_kws,linewidth=4,linecolor="k")
```

```python
plt.figure(figsize=(16,9))
annot_kws={"fontsize":10,"fontstyle":"italic","color":"k",
           "alpha":0.9,
           "backgroundcolor":"w"
           
           
          }
sns.heatmap(df_1,vmin=False,cmap="coolwarm",annot=True,annot_kws=annot_kws,linewidth=4,linecolor="k",cbar=False) # use of cbar for hide of line
```

```python
plt.figure(figsize=(16,9))
annot_kws={"fontsize":10,"fontstyle":"italic","color":"k",
           "alpha":0.9,
           "backgroundcolor":"w"
          }
ax=sns.heatmap(df_1,vmin=False,cmap="coolwarm",annot=True,annot_kws=annot_kws,linewidth=4,linecolor="k",xticklabels=np.arange(0,15))
ax.set(title="global_warming",
       ylabel="country_name",
       xlabel="year")
# use of font_scale for chage of font size
sns.set(font_scale=3)
```
