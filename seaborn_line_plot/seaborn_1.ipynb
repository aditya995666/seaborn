```python
temp=pd.DataFrame({"day":day,"temperature":temperature})
sns.lineplot(x="day",y="temperature",data=temp)
plt.show()
```

```python
# load dataseet
tips_df=sns.load_dataset("tips")
tips_df.head()
```

```python
line_plot=sns.lineplot(x="tip",y="total_bill",data=tips_df)
plt.show()
```

```python
line_plot=sns.lineplot(x="size",y="total_bill",data=tips_df)
plt.show()
```

```python
line_plot=sns.lineplot(x="day",y="total_bill",data=tips_df)
plt.show()
```

```python
line_plot=sns.lineplot(x="size",y="total_bill",data=tips_df,hue="sex")
plt.show()
```

```python
line_plot=sns.lineplot(x="size",y="total_bill",data=tips_df,hue="sex",style="sex")
plt.show()
```

```python
line_plot=sns.lineplot(x="size",y="total_bill",data=tips_df,hue="sex",style="sex",markers=["o","<"])
plt.show()
```

```python
line_plot=sns.lineplot(x="size",y="total_bill",data=tips_df,hue="sex",style="sex",markers=["o","<"],palette="hot")
plt.show()
```

```python

line_plot=sns.lineplot(x="size",y="total_bill",data=tips_df,hue="sex",style="sex",markers=["o","<"],palette="hot")
plt.xlabel("size",fontsize=15)
plt.ylabel("total_bill",fontsize=15)
plt.title("tips_data_frame",fontsize=15)
plt.show()
```

```python
line_plot=sns.lineplot(x="size",y="total_bill",data=tips_df,hue="day",style="day",markers=["o","<",">","^"],palette="hot")
plt.xlabel("size",fontsize=15)
plt.ylabel("total_bill",fontsize=15)
plt.title("tips_data_frame",fontsize=15)
plt.show()
```
