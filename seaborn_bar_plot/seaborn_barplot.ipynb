```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

how to plot seaborn barplot

```python
df=sns.load_dataset("tips")
df
```

```python
sns.barplot(x=df.total_bill,y=df.day,hue=df.sex)
```

```python
sns.barplot(x="total_bill",y="day",hue="sex",data=df) # use of hue parmeter compare of male and female data
```

```python
order=["sun","sat","fri","thur"]
sns.barplot(x="day",y="total_bill",data=df,hue="sex",order=order) 
```

```python
hue_order=["male","female"]
sns.barplot(x="day",y="total_bill",data=df,hue="sex")
```

```python
sns.barplot(x="day",y="total_bill",data=df,hue="sex",estimator=np.max)
```

```python
sns.barplot(x="day",y="total_bill",data=df,hue="sex",estimator=np.mean,) # usse of estimater mean and max value ke according ploot kerte hy
```

```python
sns.barplot(x="day",y="total_bill",data=df,hue="sex",estimator=np.mean,    n_boot=5000,) # usse of estimater mean and max value ke according ploot kerte hy
```

```python
sns.barplot(y="day",x="total_bill",data=df,hue="sex",estimator=np.mean, n_boot=5000,orient="h") # usse of estimater mean and max value ke according ploot kerte hy
```

```python
sns.barplot(x="day",y="total_bill",data=df,hue="sex",estimator=np.mean,    n_boot=5000,color="g")
           
```

```python
sns.barplot(x="day",y="total_bill",data=df,hue="sex",estimator=np.mean, palette="hot")
```

```python
sns.barplot(x="day",y="total_bill",data=df,hue="sex",estimator=np.mean, palette="hot",saturation=2)
```

```python
sns.barplot(x="day",y="total_bill",data=df,hue="sex",estimator=np.mean, palette="hot",saturation=2,fill=False)
```

```python
sns.barplot(x="day",y="total_bill",data=df,hue="sex",estimator=np.mean, palette="hot",saturation=2,width=2)
```

```python
  sns.barplot(x="day",y="total_bill",data=df,hue="sex",estimator=np.mean, palette="hot",saturation=2,ci=0.2) # use of ci change line height
```

```python
  sns.barplot(x="day",y="total_bill",data=df,hue="sex",estimator=np.mean, palette="hot",saturation=2) 
```

```python
kwargs={"alpha":0.9,}
sns.barplot(x="day",y="total_bill",data=df,hue="sex",estimator=np.mean, palette="hot",saturation=2,**kwargs)  # use of key woords orgument
```

```python
sns.set()
kwargs={"alpha":0.9,"linewidth":6,"linestyle":":","edgecolor":"k"}
sns.barplot(x="day",y="total_bill",data=df,hue="sex",estimator=np.mean, palette="hot",saturation=2,**kwargs)  # use of key woords orgument
```

```python
sns.set()
kwargs={"alpha":0.9,"linewidth":6,"linestyle":":","edgecolor":"k"}
sns.barplot(x="day",y="total_bill",data=df,hue="sex",estimator=np.mean, palette="hot",saturation=2,**kwargs)
plt.title("restorent_bill",fontsize=20)
plt.xlabel("day")
plt.ylabel("total_bill")
plt.show()
```

```python
sns.set()
plt.figure(figsize=(16,9))
kwargs={"alpha":0.9,"linewidth":6,"linestyle":":","edgecolor":"k"}
sns.barplot(x="day",y="total_bill",data=df,hue="sex",estimator=np.mean, palette="hot",saturation=2,**kwargs)
plt.title("restorent_bill",fontsize=20)
plt.xlabel("day")
plt.ylabel("total_bill")
plt.show()
```
