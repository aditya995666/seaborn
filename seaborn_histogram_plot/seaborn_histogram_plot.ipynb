```python
tips_df=sns.load_dataset("tips")
tips_df
```

```python
# plot histogram
sns.histplot(tips_df["total_bill"],kde=True)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"],color="k")
plt.legend("bill")
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set()
sns.histplot(tips_df["total_bill"],color="k")
plt.legend("bill")
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set()
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
sns.histplot(tips_df["total_bill"],color="k",bins=bins)
plt.xticks(bins)
plt.legend("bill")
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set()
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
sns.distplot(tips_df["total_bill"],color="k",bins=bins)
plt.xticks(bins)  # use of bins and xticks function for create a proper wave
plt.legend("bill")
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"])
plt.xticks(bins)
plt.legend("bill")
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.histplot(tips_df["total_bill"],color="k",kde=True,stat='count',thresh=0.1,)
sns.histplot(tips_df["size"],color="b",kde=True)
sns.histplot(tips_df["tip"],color="r",kde=True,)
plt.legend("bill")
plt.title("total restorent bill",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.histplot(tips_df["total_bill"],color="k",kde=True,stat='count',thresh=0.1,discrete=True,) # use descrrette function
sns.histplot(tips_df["size"],color="b",kde=True,discrete=True)
sns.histplot(tips_df["tip"],color="r",kde=True,discrete=True)
plt.legend("bill")
plt.title("total restorent bill",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.histplot(tips_df["total_bill"],color="k",kde=True,stat='count',thresh=0.1,discrete=True,fill=False) # use descrrette function
sns.histplot(tips_df["size"],color="b",kde=True,discrete=True,fill=True)
sns.histplot(tips_df["tip"],color="r",kde=True,discrete=True,fill=False)
plt.legend("bill")
plt.title("total restorent bill",fontsize=20)
plt.show()
```
